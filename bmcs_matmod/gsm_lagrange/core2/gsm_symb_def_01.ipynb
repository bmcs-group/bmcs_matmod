# GSM Symbolic Definition - Vector Demonstration

This notebook demonstrates the `eps_a` and `Eps_a` properties in GSMSymbDef models, showing how external and internal strain variables are collected into homogeneous vector structures.

**Models Tested:**
- 1D Models: GSM1D_ED, GSM1D_VE, GSM1D_EPD
- 3D Interface Model: GSM3D_IFC_ED

```python
# Import and Setup - with fresh module loading
import sympy as sp
import sys
from pathlib import Path

# Clear any cached modules
modules_to_clear = ['gsm_symb_def', 'gsm1d', 'gsm3d_ifc']
for module in modules_to_clear:
    if module in sys.modules:
        del sys.modules[module]

# Add core2 to path (current directory)
sys.path.insert(0, str(Path.cwd()))

# Fresh imports - use the GSMSymbDef models from current directory
from gsm_symb_def import GSMSymbDef
from gsm1d import GSM1D_ED, GSM1D_VE, GSM1D_EPD

from IPython.display import display, Math

# Enable pretty printing
sp.init_printing()

print("GSM Vector Structure Demonstration")
print("=" * 45)
print("Testing eps_a and Eps_b properties with GSMSymbDef models from core2")
```

```python
# Create all models and test what works
# Note: Using the GSMSymbDef models from core2 directory
models = {
    'ED (1D)': GSM1D_ED(),
    'VE (1D)': GSM1D_VE(), 
    'EPD (1D)': GSM1D_EPD(),
}

print("Testing GSMSymbDef model classes from core2:")
print("=" * 50)

for name, model in models.items():
    print(f"\n{name} Model:")
    
    # Display variable definitions
    print(f"  eps_vars: {model.eps_vars}")
    print(f"  Eps_vars: {model.Eps_vars}")
    print(f"  Sig_vars: {model.Sig_vars}")
    print(f"  Sig_signs: {model.Sig_signs}")
    
    # Test the vector properties
    print(f"  eps_a: {model.eps_a} (shape: {model.eps_a.shape})")
    print(f"  Eps_B: {model.Eps_B} (shape: {model.Eps_B.shape})")
    print(f"  Eps_b: {model.Eps_b} (shape: {model.Eps_b.shape})")
    print(f"  Sig_B: {model.Sig_B} (shape: {model.Sig_B.shape})")
    print(f"  Sig_b: {model.Sig_b} (shape: {model.Sig_b.shape})")

    # Display LaTeX for external strains
    display(Math(f"\\varepsilon_a = {sp.latex(model.eps_a)}"))
    # Display LaTeX for internal strains (block form)
    display(Math(f"\\mathcal{{E}}_B = {sp.latex(model.Eps_B)}"))
    # Display LaTeX for internal strains (flat form)
    display(Math(f"\\mathcal{{E}}_b = {sp.latex(model.Eps_b)}"))
    # Display LaTeX for thermodynamic forces (block form)
    display(Math(f"\\mathcal{{S}}_B = {sp.latex(model.Sig_B)}"))
    # Display LaTeX for thermodynamic forces (flat form)
    display(Math(f"\\mathcal{{S}}_b = {sp.latex(model.Sig_b)}"))
    print("-" * 40)
```

```python
# Test a specific model - GSM1D_ED
ed_model = GSM1D_ED()

print("GSM1D_ED Model Analysis:")
print("=" * 30)

print(f"Free Energy: F = {ed_model.F_expr}")
print(f"Dissipation: f = {ed_model.f_expr}")

# Test derived expressions
print(f"\nDerived expressions:")
print(f"sig_a_ = {ed_model.sig_a_}")
print(f"dF_dEps_b = {ed_model.dF_dEps_b}")
print(f"Y_b = {ed_model.Y_b}")
print(f"Sig_b_ = {ed_model.Sig_b_}")

# Display validation result
validation = ed_model.validate_symbolic_expressions()
print(f"\nModel validation: {'✓ Valid' if validation.is_valid else '✗ Issues found'}")
if not validation.is_valid:
    for issue in validation.issues:
        print(f"  - {issue}")
```

```python
# Display the GSM1D_ED model expressions in LaTeX
ed_model = GSM1D_ED()

print("GSM1D_ED - Elastic Damage Model")
print("=" * 35)

print("Free Energy Expression:")
display(Math(f"F = {sp.latex(ed_model.F_expr)}"))

print("Dissipation Criterion:")
display(Math(f"f = {sp.latex(ed_model.f_expr)} \\leq 0"))

print("External strain derivatives (stress):")
display(Math(f"\\sigma_a = \\frac{{\\partial F}}{{\\partial \\varepsilon_a}} = {sp.latex(ed_model.sig_a_)}"))

print("Internal variable derivatives:")
display(Math(f"\\frac{{\\partial F}}{{\\partial \\mathcal{{E}}_b}} = {sp.latex(ed_model.dF_dEps_b)}"))

print("Sign matrix:")
display(Math(f"Y_b = {sp.latex(ed_model.Y_b)}"))

print("Thermodynamic forces:")
display(Math(f"\\mathcal{{S}}_b = Y_b \\cdot \\frac{{\\partial F}}{{\\partial \\mathcal{{E}}_b}} = {sp.latex(ed_model.Sig_b_)}"))

print("Variable Summary:")
print(ed_model.get_expression_summary())
```

```python
# Test GSM1D_EPD model - more complex with multiple internal variables
epd_model = GSM1D_EPD()

print("GSM1D_EPD - Elastic Plastic Damage Model")
print("=" * 40)

print("Variables:")
print(f"  External: {epd_model.eps_vars}")
print(f"  Internal: {epd_model.Eps_vars}")
print(f"  Forces: {epd_model.Sig_vars}")
print(f"  Signs: {epd_model.Sig_signs}")

print(f"\nVector structures:")
print(f"  eps_a: {epd_model.eps_a}")
print(f"  Eps_B: {epd_model.Eps_B}")
print(f"  Eps_b: {epd_model.Eps_b}")
print(f"  Sig_B: {epd_model.Sig_B}")
print(f"  Sig_b: {epd_model.Sig_b}")

print("Free Energy Expression:")
display(Math(f"F = {sp.latex(epd_model.F_expr)}"))

print("Thermodynamic Forces:")
display(Math(f"\\mathcal{{S}}_b = {sp.latex(epd_model.Sig_b_)}"))
```
